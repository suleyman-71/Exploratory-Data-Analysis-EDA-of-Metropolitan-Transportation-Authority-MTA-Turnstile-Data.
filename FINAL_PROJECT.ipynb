```python
def make_dic(zip_column):
    zip_list = []
    zip_set_list = []
    dic = {}
    zip_list = list(zip_column)
    zip_set_list = list(set(zip_column))
    for i in zip_set_list:
        dic[i] = zip_list.count(i)
    return dic
```
